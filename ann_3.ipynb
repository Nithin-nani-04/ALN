<h1>Creating a dataset</hl>


```python
from sklearn.datasets import make_blobs
s_data=make_blobs(n_samples=1000, n_features=2, centers=2, cluster_std=2)
sData,target=s_data

print(sData.shape)
print(target.shape)
```

<h1>Implementation of code</h1>



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
```

```python
x_train,x_test,y_train,y_test=train_test_split(sData,target,test_size=0.3,random_state=101)
print(x_train.shape,x_test.shape)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
pre=Perceptron()
pre.fit(x_train,y_train)
y_pred=pre.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
x_min, x_max = x_train[:, 0].min() - 1, x_train[:, 0].max() + 1
y_min, y_max = x_train[:, 1].min() - 1, x_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))
Z = pre.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train)
plt.show()
```
