```python
from sklearn.datasets import make_blobs
s_data=make_blobs(n_samples=1000, n_features=2, centers=2, cluster_std=2)
sData,target=s_data

print(sData.shape)
print(target.shape)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
np.random.seed(43)
```

<h1 color='red'>sk_learn</h1>
the following sci-kit learn libary is used to import the important functions for classification and prediction and training of the model to obtain the **accuracy** score;

```python
x_train,x_test,y_train,y_test=train_test_split(sData,target,test_size=0.3,random_state=101)
print(x_train.shape,x_test.shape)
```

here we divided single data into train and test data form model training and prediction

```python
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

we standardize the data by mean and varience of test data

```python
clf=Perceptron()
clf.fit(x_train,y_train)
pred=clf.predict(x_test)
acc=np.mean(pred==y_test)
print(acc)

#Accu=accuracy_score(pred,y_test)
#print(Accu)
```

Accuracy of the modell is 0.95, it can be increased by decreasing the centers and clu_std in make_blobs
